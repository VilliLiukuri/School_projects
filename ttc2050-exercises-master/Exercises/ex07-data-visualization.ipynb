1 Read the earthquakes.csv file and save its content into a variable. Then use bar chart to visualize magnitude values being 8.2 or over so that date values will be presented in x-axle and magnitude in y-axle.

```python
import pandas as pd
earthquakes_data = pd.read_csv("data_files/earthquakes.csv",delimiter=",")
filter_data = earthquakes_data[(earthquakes_data["Magnitude"] >= 8.2)]
filter_data.plot(kind="bar",x="Date",y="Magnitude",figsize=(25,5))
```

2 Filter earthquake data so that all occurrances between time values 10:00 - 12:00 will be visualized with line chart. Group this data so that the average (mean) depth values will be presented for each source.

```python
filter_data1 = earthquakes_data[(earthquakes_data["Time"] >= "10:00") & (earthquakes_data["Time"] <= "12:00")]
filter_data2 = filter_data1.groupby("Source")["Depth"].mean()
filter_data2.plot(kind="line",x="Time",y=["Source","Depth"],figsize=(25,5))
```

gamedata = pd.read_json("data_files/gamedata.json")3 Read the gamedata.json file to a new variable. Visualize the data with pie chart so that share count will be counted for each Steam rating (in other words how many percentage out of all games have received the rating of *Very Positive*).

```python
gamedata = pd.read_json("data_files/gamedata.json")
gamedata1 = gamedata.groupby("steamRatingText").count()
label_lst =  ["{} %".format(round(i/len(gamedata)*100,1)) for i in gamedata["steamRatingPercent"]] 
gamedata1.plot(kind="pie",y="steamRatingPercent",labels=label_lst,figsize=(10,10)).legend(gamedata["steamRatingText"])
```
